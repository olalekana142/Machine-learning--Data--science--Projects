```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df= df.drop(['customerID'], axis=1)
```

```python
df
```

```python
df.drop(columns= ['TotalCharges'], axis=1, inplace=True)
```

```python
df
```

```python
nom_col= [0,6,7,9,16]
ordinal_col= [2,3,5,8,10,11,12,13,14,15]
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import  make_column_transformer
from sklearn import set_config
trans = make_column_transformer((OneHotEncoder(sparse=False),nom_col),
                               (OrdinalEncoder(), ordinal_col),
remainder ='passthrough')
set_config(display='diagram')
```

```python
trans
```

```python
x= df.iloc[:, 0:18]
y= df.Churn
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y, test_size=0.35)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
KNN = KNeighborsClassifier(9)
pipe_KNN = make_pipeline(trans, KNN)
```

```python
KNN
```

```python
pipe_KNN
```

```python
pipe_KNN.fit(x_train,y_train)
```

```python
pred_KNN= pipe_KNN.predict(x_test)
```

```python
pred_KNN
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(pred_KNN, y_test)*100
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(pipe_KNN, x_test, y_test)
```

```python
from imblearn.over_sampling import RandomOverSampler
over = RandomOverSampler()
u_x,u_y = over.fit_resample(x,y)
u_y.value_counts()
```

```python
xtrain, xtest, ytrain,ytest = train_test_split(u_x,u_y, test_size=0.35)
```

```python
pipe_KNN.fit(xtrain,ytrain)
```

```python
preKNN= pipe_KNN.predict(xtest)
```

```python
accuracy_score(preKNN, ytest)
```

```python
plot_confusion_matrix(pipe_KNN, xtest, ytest)
```
